```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
PCA_full=pd.read_csv("C:\\Users\\greco\\OneDrive\\Área de Trabalho\\OneDrive\\Glaucio\FBPase\\PCA\\PCA_FBPase_full_results_proj.csv")
lig_188=PCA_full.loc[1:1005:1]
lig_189=PCA_full.loc[1006:2010:1]



plot = sns.jointplot(x="PC_1", y="PC_2", data= lig_188, color = "silver", kind="kde", fill=True)


plot.ax_marg_x.set_xlim(-25, 20)
plot.ax_marg_y.set_ylim(-20, 15)

flatui = ["#4E8AA3", "#67C7F0","#F0CB5B","#582EA3","#894FF0","#A36258","#F0D83A","#7DF0D4","#1A5CA3","#F044BD","#59A39B"]


fig1 = plt.gcf()
plt.rcParams['axes.labelsize'] = 24
plt.rcParams['axes.titlesize'] = 24
plt.xlabel("PC1 (29.3%)")
plt.ylabel("PC2 (17.7%)")
plt.show()
plt.draw()
fig1.savefig('188.pdf', dpi=600)
```
